```python
# Imports

from tourny.structure.foundation import *
from tourny.structure.teams import *
from tourny.analysis.team import *
from tourny.analysis.tournament import *
from tourny.visualization.tournament import *
from tourny.visualization.analysis import *
```

```python
# Exploring leagues

countTeams(quidditch), countTeams(chess), countTeams(avengers), countTeams(hockey)
```

```python
# Choosing a leagues

league = quidditch
# league = chess
# league = avengers
# league = hockey
```

```python
# Making a tournament

tourny1 = Tournament(league, 'random')

print(getTeamNames(tourny1))
```

```python
# Best team

bestTeam = None
for team in getBestTeams(league):
    bestTeam = team.name
    print(team)
```

```python
# Running a tournament

tourny1 = Tournament(league, 'random')
tourny1.runTournament()

for tournyRound in tourny1.rounds:
    print(getRoundWinnerNames(tournyRound))
```

```python
# Visualizing a tournament

treeTournament(tourny1)
```

```python
# Error handling

dummyTeam1 = Team('Team1', 0)
dummyTeam2 = Team('Team2', 0)

dummyTeam1.probabilityOfWinning(dummyTeam2)
```

```python
# Winner probabilities by pairing type

pairingTypes = ['similar', 'random', 'opposite']

for pairingType in pairingTypes:
    
    tourny = Tournament(league, pairingType)
    
    winnerProbabilities = getWinnerProbabilities(tourny)
    
    print(pairingType, winnerProbabilities)
```

```python
# Round probabilities by pairing type

for pairingType in pairingTypes:
    
    tourny = Tournament(league, pairingType)
    
    roundProbabilities = getRoundProbablities(tourny, bestTeam)
    
    print(pairingType, roundProbabilities)
```
